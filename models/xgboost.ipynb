# Section 1: processing dataset for  hyperparameter tuning

```python
import numpy as np 
from sklearn.utils import resample
from sklearn.model_selection import train_test_split 
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,auc,roc_curve
```

```python
from sklearn.utils import resample
def make_balance(fit_df):
    fit_df_0=fit_df[fit_df[:,-1]==0]
    fit_df_1=fit_df[fit_df[:,-1]==1]
    max_count=np.floor(len(fit_df_0)/len(fit_df_1))
    residuals=int(len(fit_df_0)-max_count*len(fit_df_1))
    balance_count=1
    fit_df2=np.array(pd.DataFrame(fit_df_1).copy())
    while balance_count<max_count:
        fit_df2=np.concatenate((fit_df2,fit_df_1))
        balance_count=balance_count+1
    fit_df2=np.concatenate((fit_df2,resample(fit_df_1,replace=False,n_samples=residuals)))
    fit_df2=np.concatenate((fit_df2,fit_df_0))
    fit_df2=resample(fit_df2,replace=False,n_samples=len(fit_df2))
    return fit_df2
```

```python
def xgb_val(train_np,params,test_size,fold_number):

    tree_method=params['tree_method']
    objective_function=params['objective_function']
    num_trees_vec=params['n_estimators']
    depth_vec=params['max_depth']
    eta_vec=params['eta']

    prediction_dict={}
    count=1

    for n_estimators in num_trees_vec:
        fit_np,val_np=train_test_split(train_np,random_state=None,test_size=test_size,stratify=train_np[:,-1])

        fit_np=make_balance(fit_np) 
        X_fit=fit_np[:,:-1]
        y_fit=fit_np[:,-1]

        X_val=val_np[:,:-1]
        y_val=val_np[:,-1]

        for depth in depth_vec:
            for eta in eta_vec:
                model=XGBClassifier(objective=objective_function,           
                                    n_jobs=-1,tree_method=tree_method,
                                    n_estimators=n_estimators,max_depth=depth,learning_rate=eta)
                model.fit(X_fit,y_fit)  
                pred=model.predict(X_val)
                auc_score=roc_auc_score(y_val,pred)
                prediction_dict['model_{}_fold_number_{}'.format(count,fold_number)]=[n_estimators,depth,eta,auc_score]
                print(prediction_dict['model_{}_fold_number_{}'.format(count,fold_number)])
                count=count+1

    prediction_dict=pd.DataFrame(prediction_dict)
    prediction_dict.index=pd.Series(['n_estimators','max_depth','learning_rate','auc'])
    prediction_dict=prediction_dict.transpose()
    return prediction_dict.sort_values('auc',ascending=False)
```

```python
def all_models(train_np,params,test_size,resampling_instances):
    
    model_agg_df=pd.DataFrame()
    
    for i in range(resampling_instances):
        model=xgb_val(train_np,params,test_size,i+1)
        model_agg_df=pd.concat([model_agg_df,model])

    return model_agg_df
```

```python
params={'n_estimators':list(pd.Series(np.round(np.linspace(50,500,20))).astype(int)),'max_depth':[3,5,8],'eta':[0.1,0.05,0.01],'tree_method':'hist','objective_function':'binary:logistic'}
```

```python
model_agg=all_models(np.array(train_df),params,0.3,20)
```

```python
model_summary=model_agg.groupby(['n_estimators','max_depth','learning_rate']).agg({'auc':'mean'}).reset_index()
model_summary.columns=['n_estimators','max_depth','learning_rate','mean_auc']
model_summary.sort_values('mean_auc',ascending=False,inplace=True)
model_summary.head(20)
```

```python
train_np=make_balance(np.array(train_df))
X_train=train_np[:,:-1]
y_train=train_np[:,-1]
model=XGBClassifier(objective='binary:logistic',n_jobs=-1,tree_method='hist',
                                    n_estimators=335,max_depth=3,learning_rate=0.05).fit(X_train,y_train)
test_np=np.array(test_df)
X_test=test_np[:,:-1]
y_test=test_np[:,-1]
pred=model.predict(X_test)
roc_auc_score(y_test,pred)
```

```python
#define metrics
import matplotlib.pyplot as plt
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr)
plt.plot(fpr,fpr,'--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
